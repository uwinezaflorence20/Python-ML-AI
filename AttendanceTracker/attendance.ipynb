```python
# import openpyxl
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

FILE_PATH   = r'C:\Users\user\OneDrive\Desktop\attendance.xlsx'
FROM_EMAIL  = 'fillette2381@gmail.com'
FROM_PASS   = 'uewwkhstvcckakby'
STAFF_MAILS = ['uwinezaflorence20@gmail.com',
               'munezeroernest250@gmail.com',
               'uwimbabaziadelphine20@gmail.com']

book  = openpyxl.load_workbook(FILE_PATH)
sheet = book['Sheet1']
r     = sheet.max_row

m1 = "Warning! You can take only ONE more leave for CI class."
m2 = "Warning! You can take only ONE more leave for Python class."
m3 = "Warning! You can take only ONE more leave for DM class."

def savefile():
    book.save(FILE_PATH)
    print("✅ Excel saved!")

def mailstu(li, msg):
    try:
        s = smtplib.SMTP('smtp.gmail.com', 587, timeout=120)
        s.starttls()
        s.login(FROM_EMAIL, FROM_PASS)
        for to_id in li:
            message = MIMEMultipart()
            message['Subject'] = 'Attendance Report'
            message.attach(MIMEText(msg, 'plain'))
            s.sendmail(FROM_EMAIL, to_id, message.as_string())
        s.quit()
        print("📧 Mail sent to student(s)!")
    except Exception as e:
        print("❌ Failed to send student mail:", e)

def mailstaff(mail_id, msg):
    try:
        s = smtplib.SMTP('smtp.gmail.com', 587, timeout=120)
        s.starttls()
        s.login(FROM_EMAIL, FROM_PASS)
        message = MIMEMultipart()
        message['Subject'] = 'Lack of Attendance Report'
        message.attach(MIMEText(msg, 'plain'))
        s.sendmail(FROM_EMAIL, mail_id, message.as_string())
        s.quit()
        print("📧 Mail sent to staff!")
    except Exception as e:
        print("❌ Failed to send staff mail:", e)

def check(no_of_days, row_num, b):
    l1 = []
    l2 = ""
    l3 = []
    subject  = ["CI", "Python", "Data Mining"][b - 1]
    warnings = [m1, m2, m3]
    for i in range(len(row_num)):
        if no_of_days[i] == 2:
            email = sheet.cell(row=row_num[i], column=2).value
            l1.append(email)
        elif no_of_days[i] > 2:
            roll  = str(sheet.cell(row=row_num[i], column=1).value)
            email = sheet.cell(row=row_num[i], column=2).value
            l2   += roll + " "
            l3.append(email)
    if l1:
        mailstu(l1, warnings[b - 1])
    if l2 and l3:
        mailstu(l3, f"❌ You have lack of attendance in {subject}!")
        mailstaff(
            STAFF_MAILS[b - 1],
            f"The following students have lack of attendance in {subject}: {l2}"
        )

resp = 1
while resp == 1:
    print("\n1 --> CI")
    print("2 --> Python")
    print("3 --> DM")
    y = int(input("Enter subject number (1/2/3): "))
    n = int(input("Number of absent students: "))
    if n == 1:
        x = [int(input("Enter roll number: "))]
    else:
        x = list(map(int, input("Enter roll numbers separated by space: ").split()))
    col = y + 2
    row_num    = []
    no_of_days = []
    for student in x:
        for i in range(1, r + 1):
            if sheet.cell(row=i, column=1).value == student:
                current = sheet.cell(row=i, column=col).value
                updated = current + 1
                sheet.cell(row=i, column=col).value = updated
                savefile()
                no_of_days.append(updated)
                row_num.append(i)
                break
    check(no_of_days, row_num, y)
    resp = int(input("\nAnother subject? 1 = Yes   0 = No: "))

print("\n✅ All done! Excel has been updated.")
```
