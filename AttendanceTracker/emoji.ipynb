```python
import demoji
# Download emoji codes (only needed once)
demoji.download_codes()

# Example 1: Remove emojis from text
text = "Good morning! ☀️🌸"
a1 = demoji.replace(text, "")
print("Example 1 Output:", a1)

# Example 2: Replace emojis with description
text = "Happy birthday! 🎂🎈"
a2 = demoji.replace_with_desc(text)
print("Example 2 Output:", a2)

# Example 3: Detect all emojis in a string
text = "Good morning ☀️😊"
emojis = demoji.findall(text)
print("Example 3 Output:", emojis)

# Example 4: Remove all emojis
text = "Welcome to the party! 🎉🥳"
a4 = demoji.replace(text, "")
print("Example 4 Output:", a4)

# Example 1 Output: Good morning!
# Example 2 Output: Happy birthday! :birthday cake::balloon:
# Example 3 Output: {'☀️': 'sun', '😊': 'smiling face with smiling eyes'}
# Example 4 Output: Welcome to the party!
```
